# とうとうディープラーニング

- tensorflow の中の keras を使ってみます

```python
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import SGD
```

- ランダムシードを固定しておきます

```python
np.random.seed(0)
```

### こういうのを Keras で実装してみましょう

- コードと図をよく見比べてみてください。そんなに難しくはないと思います。
#### このサイト [「入門 Keras (3) 線形分離と多層パーセプトロン」](https://www.infiniteloop.co.jp/blog/2017/12/learning-keras-03/) を参考にしました。

```python
model = Sequential()
model.add(Dense(input_dim=2, units=2))
model.add(Activation('sigmoid'))
model.add(Dense(units=1))
model.add(Activation('sigmoid'))
model.compile(loss='binary_crossentropy',optimizer=SGD(lr=0.1))
```

- 排他的OR を作るべく、学習用データを作ります

```python
X = np.array( [ [0,0],[0,1],[1,0],[1,1] ])  # 入力
T = np.array( [ [0],[1],[1],[0] ])             # 排他的OR出力
```

- 1500回、繰り返し学習して、重み係数を求めていきます

```python
model.fit(X,T, epochs=1500, batch_size=1)
```

- 損失関数の値が、学習を繰り返すごとに小さくなっているのがわかります

- 学習用とテスト用が同じですが、評価してみましょう

```python
Y = model.predict_classes(X, batch_size=1)
```

```python
print(Y)
```

```python
print(Y==T)
```
