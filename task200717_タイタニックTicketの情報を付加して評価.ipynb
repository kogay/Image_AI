```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head(20)
```

```python
df.tail()
```

```python
df.info()
```

```python
tmp = df['Ticket'].str.split( )
tmp
```

```python
tmp = df['Ticket'].str.extract('(.*?) ?([0-9]+)$')
tmp
```

```python
df['Ticket_pfx'] = tmp[0]
df['Ticket_no'] = tmp[1]
df.head()
```

```python
df['Ticket'].str.len()
```

```python
df['Ticket_len'] = df['Ticket'].str.len()
```

```python
df.head()
```

```python
df['Ticket'].str.get(0)
```

```python
df['Ticket_init'] = df['Ticket'].str.get(0)
```

```python
df = df.drop(['Ticket_pfx','Ticket_no'], axis=1)
```

```python
df.head()
```

```python
type(df.loc[4,'Ticket_init'])
```

```python
df['Ticket_init'].value_counts()
```

```python
df_copy = df.copy()
df_copy['Ticket_init'] = df['Ticket_init'].map({'3':0,'2':1,'1':2,'S':3,'P':4,'C':5,'A':6,'W':7,'4':8,'7':9,'F':10,'6':11,'L':12,'5':13,'8':14,'9':15})
```

```python
df_copy.head()
```

```python
df_copy['Sex'] = df['Sex'].map({'male':0, 'female':1})
df_copy.head()
```

```python
df_copy.isnull().sum()
```

```python
df_copy['Age'] = df['Age'].fillna(df['Age'].mean())
df_copy['Embarked'] = df['Embarked'].fillna('S')
```

```python
df_copy['Embarked'].unique()
```

```python
df_copy['Embarked'] = df_copy['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
df_copy.info()
```

```python
data_org = df_copy[['Pclass','Sex','Age','Fare','Embarked']]
data = df_copy[['Pclass','Sex','Age','Fare','Embarked','Ticket_len','Ticket_init']]
label = df_copy['Survived']
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
train_d1, test_d1, train_l1, test_l1 = train_test_split(data_org, label, test_size=0.3, random_state=0)
train_d2, test_d2, train_l2, test_l2 = train_test_split(data, label, test_size=0.3, random_state=0)
```

```python
clf1 = SVC(C=1, gamma=0.1)
clf2 = SVC(C=1, gamma=0.1)
scores1 = cross_val_score(clf1, data_org, label, cv=5)
scores2 = cross_val_score(clf2, data, label, cv=5)
```

```python
score1 = np.mean(scores1)
score2 = np.mean(scores2)
print("SVMでの評価")
print(f"original:\t\t{score1:.3f}")
print(f"add ticket info:\t{score2:.3f}")
```

```python
print(scores1)
print(scores2)
```

```python
clf3 = RFC()
clf4 = RFC()
clf3.fit(train_d1, train_l1)
clf4.fit(train_d2, train_l2)
score3 = clf3.score(test_d1, test_l1)
score4 = clf4.score(test_d2, test_l2)
```

```python
print("RFCでの評価")
print(f"original:\t\t{score3:.5f}")
print(f"add ticket info:\t{score4:.5f}")
```
