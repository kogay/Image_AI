```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('train.csv')
df.info()
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
df['Sex'].head()
```

```python
label = df['Survived']
```

```python
train_d_org, test_d_org, train_l, test_l = train_test_split(df, label, test_size=0.3)
```

```python
train_d = train_d_org[['Pclass', 'Sex', 'Fare']]
test_d= test_d_org[['Pclass', 'Sex', 'Fare']]
```

```python
clf = SVC(C=10, gamma=10)
clf.fit(train_d, train_l)
pred = clf.predict(test_d)
```

```python
clf.score(test_d, test_l)
```

```python
pred_s = pd.Series(pred, index=test_l.index)
pred_s
```

```python
test_l
```

```python
print(test_l!=pred_s)
```

予測が外れたものだけ抽出

```python
test_d_f = test_d_org[test_l!=pred_s]
test_d_f
```

```python
test_d_f.to_csv('test_d_f.csv', sep='\t', index=False)
```
