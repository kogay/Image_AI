```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('train.csv')
df.info()
```

```python
df.head()
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].value_counts().index[0]
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
df['Embarked'] = df['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
df.info()
```

```python
df.corr()
```

```python
data = df.copy()[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]
label = df.copy()['Survived']
```

```python
train_d, test_d, train_l, test_l = train_test_split(data, label, test_size=0.3)
```

```python
#param_grid = {'kernel': ['rbf', 'linear'], 'C': [0.01,0.1,1,5,10,20,50], 'gamma': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1]}
param_grid = {'C': [0.01,0.1,1,5,10,20,50], 'gamma': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1]}

clf = GridSearchCV(SVC(), param_grid, cv=5)
clf.fit(train_d, train_l)
```

```python
res = pd.DataFrame(clf.cv_results_)
```

```python
res.head()
```

```python
res[res['rank_test_score'] <= 5].sort_values('rank_test_score', axis=0)
```

```python
best_params = res.loc[38].params
```

```python
clf = SVC(C=best_params['C'], gamma=best_params['gamma'])
m_score = []
for i in np.arange(10):
    train_d, test_d, train_l, test_l = train_test_split(data, label, test_size=0.3, random_state=i)
    clf.fit(train_d, train_l)
    score = clf.score(test_d, test_l)
    print(f'Score[{i}]\t\t{score:.3f}')
    m_score.append(score)

print(f'\nMean score: \t{np.mean(m_score):.3f}')
print(f'Std: \t{np.std(m_score):.3f}')
```

Stdが小さいほど、汎用性の高いモデルと言える。
