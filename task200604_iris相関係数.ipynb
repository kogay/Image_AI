```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("iris_dataset.csv")
df
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.scatter(x=df.SepalLength, y=df.SepalWidth)
plt.xlabel("SepalLength")
plt.ylabel("SepalWidth")
plt.xlim(0,10)
plt.ylim(0,10)
plt.subplot(2,3,4)
plt.scatter(x=df.SepalLength, y=df.PetalLength)
plt.xlabel("SepalLength")
plt.ylabel("PetalLength")
plt.xlim(0,10)
plt.ylim(0,10)
plt.subplot(2,3,5)
plt.scatter(x=df.SepalLength, y=df.PetalWidth)
plt.xlabel("SepalLength")
plt.ylabel("PetalWidth")
plt.xlim(0,10)
plt.ylim(0,10)
plt.subplot(2,3,6)
plt.scatter(x=df.PetalLength, y=df.PetalWidth)
plt.xlabel("PetalLength")
plt.ylabel("PetalWidth")
plt.xlim(0,10)
plt.ylim(0,10)

plt.show()
```
