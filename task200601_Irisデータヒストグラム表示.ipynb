```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('iris_dataset.csv')
```

```python
df
```

```python
df_set = df[df.Name == 'setosa']
```

```python
df_ver = df[df.Name == 'versicolor']
```

```python
df_vir = df[df.Name == 'virginica']
```

```python
df_set
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,1,1)
plt.title("Hist of Iris' PetalLength")
plt.hist(df_set.PetalLength, histtype='bar', alpha=0.5, label='setosa')
plt.hist(df_ver.PetalLength, histtype='bar', alpha=0.5, label='versicolor')
plt.hist(df_vir.PetalLength, histtype='bar', alpha=0.5, label='virginica')
plt.legend()

plt.show()
```
