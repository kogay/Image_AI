```python
avg = []
std = []
maxval = []
minval = []
for i in range(4):
    avg.append(np.round(np.average(df.iloc[:,i]), 2))
    std.append(np.round(np.std(df.iloc[:,i]), 2))
    maxval.append(np.round(np.max(df.iloc[:,i]), 2))
    minval.append(np.round(np.min(df.iloc[:,i]), 2))
```
