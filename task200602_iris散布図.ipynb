```python
plt.figure(figsize=(10,6))

plt.subplot(1,1,1)
plt.xlabel('SepalLength', fontsize=12)
plt.ylabel('SepalWidth', fontsize=12)
plt.xlim(2,10)
plt.ylim(1,5)
plt.scatter(x=a_set[:,0],y=a_set[:,1],color='r', label='setosa')
plt.scatter(x=a_ver[:,0],y=a_ver[:,1],color='g', label='versicolor')
plt.scatter(x=a_vir[:,0],y=a_vir[:,1],color='b', label='virginica')
plt.legend()
plt.show()
```
