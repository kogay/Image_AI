```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
df = pd.read_csv('wine.csv')
df.info()
```

```python
data = df.drop('label', axis=1)
label = df['label']
```

```python
train_d, test_d, train_l, test_l = train_test_split(data, label, test_size=0.3)
```

```python
param = {'C': [1e-1, 1, 10, 100, 1000, 1e4]}

clf = GridSearchCV(LogisticRegression(), param, cv=4)
clf.fit(train_d, train_l)
res = pd.DataFrame(clf.cv_results_)
res.head()
```

```python
res.sort_values(['rank_test_score', 'std_test_score'])
```

```python
clf = LogisticRegression(C=1)
clf.fit(train_d, train_l)
clf.score(test_d, test_l)
```
